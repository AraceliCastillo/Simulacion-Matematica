# Las cosas del amor.

Consideramos un modelo simple de la dinámica del amor/odio en una pareja, propuesto por Strogatz en 1988.

- Strogatz, S.H. (1988) *Love affairs and differential equations*, Math. Magazine 61, 35.

### Primer punto (50 puntos).

Romeo está enamorado de Julieta, pero en este caso, Julieta tiene sentimientos volubles. Mientras Romeo más la ama, ella quiere huir de él. Pero cuando Romeo no muestra atención, Julieta lo empieza a ver muy atractivo.

Del otro lado, Romeo tiene sentimientos imitativos. Esto quiere decir que se pone cariñoso cuando ella lo ama y se vuelve frío cuando ella lo odia.

Sean:

- $x_1(t)$ : amor/odio de Romeo hacia Julieta en el tiempo  $t$ .
- $x_2(t)$ : amor/odio de Julieta hacia Romeo en el tiempo  $t$ .

Consideraremos que un valor positivo de  $x_1$  o  $x_2$  corresponde a amor, y que un valor negativo corresponde a odio.

Un modelo del romance anteriormente descrito es
$$ \frac{dx_1}{dt}=ax_2,$$
$$ \frac{dx_2}{dt}=−bx_1$$
 
con  $a$  y  $b$  constantes positivas. Definimos  $x=[x_1,x_2]^T.

1. (20 puntos) Simular este sistema eligiendo los parámetros  $a=b=1$  y condiciones iniciales  $x(0)=[x1(0),x2(0)]^T=[5,5]^T$, usando un vector de tiempo de $0$ a $50$ (unidades de tiempo).
2. (15 puntos)
    - (10 puntos) Graficar las soluciones en el tiempo, es decir,  $x_1$  vs. $t$  y  $x_2$ vs  $t$.  
    - (5 puntos) ¿Qué forma tienen dichas soluciones? ¿Qué se puede decir del amor/odio de Romeo hacia Julieta? ¿Qué se puede decir del amor/odio de Julieta hacia Romeo?
3. (15 puntos)
    - (10 puntos) Graficar el retrato de fase, es decir,  $x_2$  vs  $x_1$.
    - (5 puntos) ¿Cómo es el amor de Romeo y Julieta? ¿Considera usted sano este tipo de amor?

```python
%matplotlib inline
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

def funcion(x,t):
    x1, x2 = x[0],x[1]
    return [a*x2, -b*x1]
a,b=1,1
t = np.linspace(0,50)
x = [5,5]

resultado = odeint(funcion,x,t)
print (resultado)
   
```

```python
def funcion(x,t):
    x1, x2 = x[0],x[1]
    return [a*x2, -b*x1]
a,b=1,1
t = np.linspace(0,50,200)
x = [5,5]

resultado = odeint(funcion,x,t)

R = resultado[:,0]
J = resultado[:,1]

plt.figure(figsize=(15,5))
plt.plot(t,R, 'g', label="R")
plt.plot(t,J,'m', label="J")
plt.title("1-ODIO Y AMOR ENTRE ROMEO Y JULIETA",fontsize = 25)
plt.xlabel("Tiempo",fontsize = 18)
plt.ylabel("Sentimientos", fontsize = 20)
plt.legend(loc = 'best', prop={'size': 16})
plt.show()
```

**Interpretación gráfica 1** 
Tienen una forma parecida a un seno a un coseno.
Después de graficar en los valores correspondientes el amor de Romeo y Julieta podemos darnos cuenta de manera visual lo que ya nos habían dicho con anterioridad (plantemiento del problema). 
En el tiempo 0 comienzan en la misma posición, amándose, debido a que es un valor positivo. Pero como Romeo la está amando Julieta comienza a odiarlo y se va hacia los valores negativos (Valores de Odio), como ella comienza a odiarlo Romeo también, pero cuando el la odia ella lo ama, así se comportan ambos a medida que va avanzando el tiempo.
Pocas veces sienten lo mismo pero siempre sienten algo por el otro.

```python
plt.figure(figsize = (6,6))
plt.scatter(J, R, lw = 5, s = 100, cmap = 'spring', c = J)
plt.title('2-Cuandrante de Fase')
plt.xlabel("Amor de Julieta a Romeo",fontsize = 18)
plt.ylabel("Amor de Romeo a Julieta", fontsize = 20)
plt.grid(True)
plt.show()
```

**Interpretación Gráfica 2**

Cuando el amor de Romeo aumenta el de Julieta disminuye

Cuando el amor de Romeo disminuye el de Julieta aumenta

Cuando el amor de Julieta aumenta el de Romeo aumenta

Cuando el amor de Julieta disminuye el de Romeo disminuye

Este amor se puede interpretar como un círculo vicioso, sin embargo, yo no lo veo completamente mal porque para que una relación funcione debe haber altibajos, si fuera siempre igual sería aburrido y podrían caer en monotomía. Sin embargo, como es un círculo vicioso uno de los dos se puede hartar y terminar la relación, en especial Romeo.
               


### Segundo punto.
En este punto consideraremos a Romeo y a Julieta como dos amantes cautelosos idénticos.

Entonces, un modelo que describe este romance es

$$\frac{dx_1}{dt} = -ax_1 + bx_2$$
$$\frac{dx_2}{dt} = bx_1 -ax_2$$
 
con  $a$  y  $b$  constantes positivas. Definimos  $x=[x_1,x_2]^T$.

Simular este sistema y obtener gráficas de  $x_1$  vs.  $t$ ,  $x_2$  vs.  $t$ y  $x_2$  vs.  $x_1$  en cada uno de los siguientes casos. Además, concluya acerca de cómo es el amor entre Romeo y Julieta cuando  $t \rightarrow \infty$.


1. (25 puntos) Para  $a=2$  y  $b=1$.

    - $x(0)=[x_1(0),x_2(0)]^T =[2,1]^T$.
    - $x(0)=[x_1(0),x_2(0)]^T =[−2,1]^T$.
    - $x(0)=[x_1(0),x_2(0)]^T =[−2,−1]^T$.
    - $x(0)=[x_1(0),x_2(0)]^T =[2,−1]^T$.
2. (25 puntos) Para  $a=1$  y  $b=2$.
    - $x(0)=[x_1(0),x_2(0)]T=[2,1]^T$.
    - $x(0)=[x_1(0),x_2(0)]T=[−2,1]^T$.
    - $x(0)=[x_1(0),x_2(0)]T=[−2,−1]^T$.
    - $x(0)=[x_1(0),x_2(0)]T=[2,−1]^T$.

```python
def funcion(x,t):
    x1, x2 = x[0],x[1]
    return [-a*x1+b*x2, b*x1-a*x2]
a,b=2,1
t = np.linspace(0,50,1000)
x = [2,1]

resultado = odeint(funcion,x,t)

R = resultado[:,0]
J = resultado[:,1]


plt.figure(figsize=(10,5))
plt.plot(t,R, 'g', label="R")
plt.plot(t,J,'m', label="J")
plt.title("ODIO Y AMOR ENTRE ROMEO Y JULIETA 1",fontsize = 15)
plt.xlabel("Tiempo",fontsize = 10)
plt.ylabel("Sentimientos", fontsize = 10)
plt.legend(loc = 'best', prop={'size': 10})
plt.show()

plt.figure(figsize = (6,6))
plt.scatter(R, J, lw = 5, s = 100, cmap = 'spring', c = J)
plt.title('1-Cuandrante de Fase')
plt.xlabel("Amor de Julieta a Romeo",fontsize = 18)
plt.ylabel("Amor de Romeo a Julieta", fontsize = 20)
plt.grid(True)
plt.show()

#-------------------------------------------------------------------------------------------------------------------------------

x1=[-2,1]
resultado1 = odeint(funcion,x1,t)
R1=resultado1[:,0]
J1=resultado1[:,1]

plt.figure(figsize=(10,5))
plt.plot(t,R1, 'g', label="R")
plt.plot(t,J1,'m', label="J")
plt.title("ODIO Y AMOR ENTRE ROMEO Y JULIETA 2 ",fontsize = 15)
plt.xlabel("Tiempo",fontsize = 10)
plt.ylabel("Sentimientos", fontsize = 10)
plt.legend(loc = 'best', prop={'size': 10})
plt.show()

plt.figure(figsize = (6,6))
plt.scatter(J1, R1, lw = 5, s = 100, cmap = 'spring', c = J)
plt.title('2-Cuandrante de Fase')
plt.xlabel("Amor de Julieta a Romeo",fontsize = 18)
plt.ylabel("Amor de Romeo a Julieta", fontsize = 20)
plt.grid(True)
plt.show()

#-------------------------------------------------------------------------------------------------------------------------------

x2=[-2,-1]
resultado2 = odeint(funcion,x2,t)
R2=resultado2[:,0]
J2=resultado2[:,1]

plt.figure(figsize=(10,5))
plt.plot(t,R2, 'g', label="R")
plt.plot(t,J2,'m', label="J")
plt.title("ODIO Y AMOR ENTRE ROMEO Y JULIETA 3",fontsize = 15)
plt.xlabel("Tiempo",fontsize = 10)
plt.ylabel("Sentimientos", fontsize = 10)
plt.legend(loc = 'best', prop={'size': 10})
plt.show()

plt.figure(figsize = (6,6))
plt.scatter(J2, R2, lw = 5, s = 100, cmap = 'spring', c = J)
plt.title('3-Cuandrante de Fase')
plt.xlabel("Amor de Julieta a Romeo",fontsize = 18)
plt.ylabel("Amor de Romeo a Julieta", fontsize = 20)
plt.grid(True)
plt.show()

#-------------------------------------------------------------------------------------------------------------------------------

x3=[2,-1]
resultado3 = odeint(funcion,x3,t)
R3=resultado3[:,0]
J3=resultado3[:,1]

plt.figure(figsize=(10,5))
plt.plot(t,R3, 'g', label="R")
plt.plot(t,J3,'m', label="J")
plt.title("ODIO Y AMOR ENTRE ROMEO Y JULIETA 4",fontsize = 15)
plt.xlabel("Tiempo",fontsize = 10)
plt.ylabel("Sentimientos", fontsize = 10)
plt.legend(loc = 'best', prop={'size': 10})
plt.show()

plt.figure(figsize = (6,6))
plt.scatter(J3, R3, lw = 5, s = 100, cmap = 'spring', c = J)
plt.title('4-Cuandrante de Fase')
plt.xlabel("Amor de Julieta a Romeo",fontsize = 18)
plt.ylabel("Amor de Romeo a Julieta", fontsize = 20)
plt.grid(True)
plt.show()
```
