```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib as mpl
label_size = 14
mpl.rcParams['xtick.labelsize'] = label_size 
mpl.rcParams['ytick.labelsize'] = label_size
```

```python
r = 1 #Velocidad a la que crecer√° 
def poblacion(x, t):
    return r * x * (1 - x)
```

```python

x0 = 0.05
tt = np.linspace(0, 10) #es lo mismo que arange, solo cambia la manera en que se distribuye
xx = odeint(poblacion, x0, tt)
plt.plot(tt, xx)
plt.xlabel('$t$', fontsize = 18)
plt.ylabel('$x$', fontsize = 18)
plt.show()
```

```python

def logi_sol(t, x0, r):
    return 1/(1 + (1/x0 - 1) * np.exp(-r * t))
```

```python
t = np.linspace(0,10)
for r in np.arange(-1, 1,.1):
    plt.plot(t, logi_sol(t, 0.05, r))
plt.xlabel('$t$', fontsize = 18)
plt.ylabel('$x$', fontsize = 18)
plt.show()
```

```python
t = np.linspace(0,10)
for r in np.arange(-1, 1,.1):
    plt.plot(t, odeint(poblacion, x0,tt))
plt.xlabel('$t$', fontsize = 18)
plt.ylabel('$x$', fontsize = 18)
plt.show()
```

```python
def funcion(x, t):
    return 2*x
```

```python
x0 = 10
tt = np.linspace(0, 10) #es lo mismo que arange, solo cambia la manera en que se distribuye
xx = odeint(funcion, x0, tt)
plt.plot(tt, xx)
plt.xlabel('$t$', fontsize = 18)
plt.ylabel('$x$', fontsize = 18)
plt.show()
```

```python
import numpy as np
def mapa_logistico(r, x):
    return r * x * (1 - x)
```

```python
r=3.7
n=100
x=np.zeros(n)
x[0]=.3

for i in range (n-1):
    x[i+1]=mapa_logistico(r,x[i])
plt.plot(np.arange(0,n),x)
```

```python
# Para mil valores de r entre 2.5 y 4.0
n = 1000
r = np.linspace(.5, 4.0, n)
```

```python
# Hacemos 1000 iteraciones y nos quedamos con las ultimas 100
iterar = 1000
ultimos = 100
```

```python
# La misma condici√≥n inicial para todos los casos. 
x = 1e-5 * np.ones(n)
```

```python
plt.figure(figsize=(7, 4.5))
for i in range(iterar):
    x = mapa_logistico(r, x)
    if i >= (iterar - ultimos):
        plt.plot(r, x, ',k', alpha=.2)
plt.xlim(.4, 4)
plt.ylim(-.1, 1.1)
plt.title("Diagrama de bifurcaci√≥n")
plt.xlabel('$r$', fontsize=18)
plt.ylabel('$x$', fontsize=18)
plt.show()
```
