```python
P=lambda t:10*np.exp(2*t) #Solución analítica

def dP (pk,h):
    pkl=pk+h*2*pk
    return pkl  #Solución Numérica

tfin=3
h=.1

t=np.arange(0,tfin+.1, .1)
pk=np.arange(0,tfin+h,h)

plt.plot(t,P(t),"b")

pk[0]=10

for k in np.arange(0,pk.size-1):  #Rango desde 0 a un número no definido
    pk[k+1]= dP(pk[k], h)
    
tk=np.arange(0,tfin+h,h)
plt.plot(tk,pk,"g")

tk= np.arange (0,tfin+h, h)
```
