```python
import matplotlib as mpl
label_size = 14
mpl.rcParams["xtick.labelsize"]=label_size
mpl.rcParams["ytick.labelsize"]=label_size
```

```python
t=np.linspace(0,50,100)
plt.figure(figsize = (10,5))
plt.plot(t, .5*np.cos(.5 * t)+ .1 * np.sin(.5*t), "y-",lw=2, ms=4, label= r'$\dot{x(t)}$')
plt.plot(t, .5*.1*np.cos(.5 * t)- .5 *.5* np.sin(.5*t), "b--",lw=3, ms=4, label= r'$\dot{x(t)}$')
plt.xlabel('$t$', fontsize= 20)
plt.show()
```

```python
a=2
w=.5
b=.1
t=np.linspace(0,50,100)
plt.figure(figsize = (10,5))
plt.plot(t, a*np.cos(w * t)+ b* np.sin(w*t), "y-",lw=2, ms=4, label= r'$\dot{x(t)}$')
plt.plot(t, w*b*np.cos(w * t)- a *w* np.sin(w*t), "b--",lw=3, ms=4, label= r'$\dot{x(t)}$')
plt.xlabel('$t$', fontsize= 20)
plt.legend(loc='best')
plt.show()
```
