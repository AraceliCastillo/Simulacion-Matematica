```python
def walker():
    global x,y
    eleccion = random.randint(0,3)
    if eleccion ==0:
        x+=1
    elif eleccion==1:
        x-=1
    elif eleccion==2:
        y+=1
    else:
        y-=1
    return [x,y]

x=0
y=0

pos = np.zeros ((2,1000))

for i in np.arange (0,len(pos[0]),1):
    pos[:,i]=walker()

plt.plot(pos[0,:], pos[1,:], 'bo-')
plt.plot(pos[0,-1], pos[1,-1],'ro-')    
```
